```python
import pandas as pd
import numpy as np
#read dataset
data = pd.read_csv(r'C:\Users\Sri\anaconda3\Breast_cancer_data.csv')

X = data.iloc[:, :-1].values


y = data.iloc[:, -1].values


np.random.seed(42)

indices = np.random.permutation(len(X))
#80 train and 20 test
split = int(0.8 * len(X))

train_index = indices[:split]
test_index = indices[split:]

X_train = X[train_index]
y_train = y[train_index]

X_test = X[test_index]
y_test = y[test_index]

classes = np.unique(y_train)

mean = {}
variance = {}
prior = {}
#Train
for c in classes:

    # rows belonging to current class
    X_c = X_train[y_train == c]

    # Calculate Mean
    mean[c] = np.mean(X_c, axis=0)

    # Calculate Variance
    variance[c] = np.var(X_c, axis=0) + 1e-9

    # Calculate Prior Probability
    prior[c] = len(X_c) / len(X_train)

def gaussian(x, mean, variance):

    numerator = np.exp(-((x - mean) ** 2) / (2 * variance))

    denominator = np.sqrt(2 * np.pi * variance)

    return numerator / denominator

#Prediction
predictions = []

for sample in X_test:

    probabilities = []

    for c in classes:

        total_probability = np.log(prior[c])

        for i in range(len(sample)):

            p = gaussian(sample[i],
                         mean[c][i],
                         variance[c][i])

            total_probability = total_probability + np.log(p)

        probabilities.append(total_probability)

    predicted_class = classes[np.argmax(probabilities)]

    predictions.append(predicted_class)

predictions = np.array(predictions)

accuracy = np.mean(predictions == y_test)

print("Predicted Values")
print(predictions)

print("\nActual Values")
print(y_test)

print("\nAccuracy =", accuracy * 100)
```

```python

predictions = np.array(predictions)


accuracy = np.mean(predictions == y_test)


classes = np.unique(y_test)

confusion_matrix = np.zeros((len(classes), len(classes)), dtype=int)

for i in range(len(y_test)):
    actual = np.where(classes == y_test[i])[0][0]
    predicted = np.where(classes == predictions[i])[0][0]
    confusion_matrix[actual][predicted] += 1

print("\nConfusion Matrix")
print(confusion_matrix)

print("\nEvaluation Metrics")

for i in range(len(classes)):

    TP = confusion_matrix[i][i]

    FP = np.sum(confusion_matrix[:, i]) - TP

    FN = np.sum(confusion_matrix[i, :]) - TP

    TN = np.sum(confusion_matrix) - TP - FP - FN

    precision = TP / (TP + FP) if (TP + FP) != 0 else 0

    recall = TP / (TP + FN) if (TP + FN) != 0 else 0

    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    print(f"\nClass : {classes[i]}")
    print(f"Precision : {precision:.4f}")
    print(f"Recall    : {recall:.4f}")
    print(f"F1 Score  : {f1:.4f}")

print(f"\nOverall Accuracy : {accuracy * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd

class NaiveBayes:

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.mean = {}
        self.var = {}
        self.prior = {}

        for c in self.classes:
            X_c = X[y == c]
            self.mean[c] = np.mean(X_c, axis=0)
            self.var[c] = np.var(X_c, axis=0) + 1e-9   # Avoid division by zero
            self.prior[c] = len(X_c) / len(X)

    def gaussian_probability(self, x, mean, var):
        exponent = np.exp(-((x - mean) ** 2) / (2 * var))
        return (1 / np.sqrt(2 * np.pi * var)) * exponent

    def predict(self, X):
        predictions = []

        for x in X:
            posteriors = []

            for c in self.classes:
                prior = np.log(self.prior[c])

                likelihood = np.sum(
                    np.log(self.gaussian_probability(x,
                                                     self.mean[c],
                                                     self.var[c]))
                )

                posterior = prior + likelihood
                posteriors.append(posterior)

            predictions.append(self.classes[np.argmax(posteriors)])

        return np.array(predictions)


# -----------------------------
# Load Dataset
# -----------------------------
data = pd.read_csv("Breast_cancer_data.csv")      # Change filename

X = data.iloc[:, :-1].values           # Features
y = data.iloc[:, -1].values            # Target

# -----------------------------
# Train-Test Split (80-20)
# -----------------------------
np.random.seed(42)

indices = np.random.permutation(len(X))
split = int(0.8 * len(X))

train_idx = indices[:split]
test_idx = indices[split:]

X_train = X[train_idx]
X_test = X[test_idx]
y_train = y[train_idx]
y_test = y[test_idx]

# -----------------------------
# Train Model
# -----------------------------
model = NaiveBayes()
model.fit(X_train, y_train)

# -----------------------------
# Prediction
# -----------------------------
predictions = model.predict(X_test)

# -----------------------------
# Accuracy
# -----------------------------
accuracy = np.mean(predictions == y_test)

print("Predictions:")
print(predictions)

print("\nActual:")
print(y_test)

print("\nAccuracy:", round(accuracy * 100, 2), "%")
```

```python
import pandas as pd
import numpy as np

# ----------------------------
# Read the dataset
# ----------------------------
data = pd.read_csv(r'C:\Users\Sri\anaconda3\Breast_cancer_data.csv')
# Features (all columns except last)
X = data.iloc[:, :-1].values

# Target (last column)
y = data.iloc[:, -1].values

# ----------------------------
# Split into Training and Testing (80-20)
# ----------------------------
np.random.seed(42)

indices = np.random.permutation(len(X))

split = int(0.8 * len(X))

train_index = indices[:split]
test_index = indices[split:]

X_train = X[train_index]
y_train = y[train_index]

X_test = X[test_index]
y_test = y[test_index]

# ----------------------------
# Find all classes
# ----------------------------
classes = np.unique(y_train)

mean = {}
variance = {}
prior = {}

# ----------------------------
# Training
# ----------------------------
for c in classes:

    # Select rows belonging to current class
    X_c = X_train[y_train == c]

    # Calculate Mean
    mean[c] = np.mean(X_c, axis=0)

    # Calculate Variance
    variance[c] = np.var(X_c, axis=0) + 1e-9

    # Calculate Prior Probability
    prior[c] = len(X_c) / len(X_train)

# ----------------------------
# Gaussian Function
# ----------------------------
def gaussian(x, mean, variance):

    numerator = np.exp(-((x - mean) ** 2) / (2 * variance))

    denominator = np.sqrt(2 * np.pi * variance)

    return numerator / denominator

# ----------------------------
# Prediction
# ----------------------------
predictions = []

for sample in X_test:

    probabilities = []

    for c in classes:

        total_probability = np.log(prior[c])

        for i in range(len(sample)):

            p = gaussian(sample[i],
                         mean[c][i],
                         variance[c][i])

            total_probability = total_probability + np.log(p)

        probabilities.append(total_probability)

    predicted_class = classes[np.argmax(probabilities)]

    predictions.append(predicted_class)

# ----------------------------
# Accuracy
# ----------------------------
predictions = np.array(predictions)

accuracy = np.mean(predictions == y_test)

print("Predicted Values")
print(predictions)

print("\nActual Values")
print(y_test)

print("\nAccuracy =", accuracy * 100)
```

```python
# ----------------------------
# Evaluation Metrics
# ----------------------------
predictions = np.array(predictions)

# Accuracy
accuracy = np.mean(predictions == y_test)

# Find unique classes
classes = np.unique(y_test)

# Confusion Matrix
confusion_matrix = np.zeros((len(classes), len(classes)), dtype=int)

for i in range(len(y_test)):
    actual = np.where(classes == y_test[i])[0][0]
    predicted = np.where(classes == predictions[i])[0][0]
    confusion_matrix[actual][predicted] += 1

print("\nConfusion Matrix")
print(confusion_matrix)

# Precision, Recall, F1 Score
print("\nEvaluation Metrics")

for i in range(len(classes)):

    TP = confusion_matrix[i][i]

    FP = np.sum(confusion_matrix[:, i]) - TP

    FN = np.sum(confusion_matrix[i, :]) - TP

    TN = np.sum(confusion_matrix) - TP - FP - FN

    # Avoid division by zero
    precision = TP / (TP + FP) if (TP + FP) != 0 else 0

    recall = TP / (TP + FN) if (TP + FN) != 0 else 0

    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    print(f"\nClass : {classes[i]}")
    print(f"Precision : {precision:.4f}")
    print(f"Recall    : {recall:.4f}")
    print(f"F1 Score  : {f1:.4f}")

print(f"\nOverall Accuracy : {accuracy * 100:.2f}%")
```
